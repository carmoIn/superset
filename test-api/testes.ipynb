```python
import requests

# Dados de acesso
superset_username = 'admin'
superset_password = 'admin'
superset_url = 'http://localhost:8088'

# Endpoint autenticação
auth_endpoint = f"{superset_url}/api/v1/security/login"

# Payload de autenticação
auth_payload = {
    "username": superset_username,
    "password": superset_password,
    "provider": "db"  # Assuming database authentication
}

# Realizando autenticação
response = requests.post(auth_endpoint, json=auth_payload)

if response.status_code == 200:
    # Obtem token de acesso
    access_token = response.json()['access_token']
    response.json()
    print("Autenticação com sucesso")
else:
    print("Falha ao autenticar")
```

```python
# SQL e dataset utilizados
sql_query = 'select * from public.channels'
datasource_id = '1'  # Replace with your datasource ID

# Endpoint para execução da query
query_endpoint = f"{superset_url}/api/v1/sqllab/execute/"

# Headers
headers = {
    'Authorization': f'Bearer {access_token}',
    'accept': 'application/json',
    'Content-Type': 'application/json'
}

# Payload for the query
query_payload = {
  "ctas_method": "string",
  "database_id": datasource_id,
  "expand_data": True,
  "json": True,
  "queryLimit": 100,
  "runAsync": False,
  "schema": "public",
  "select_as_cta": False,
  "sql": sql_query,
  "ctas_method": "TABLE",
  "sql_editor_id": "string",
  "tab": "string",
  "templateParams": "string",
  "tmp_table_name": "string"
}

# Sending a POST request to execute the query
response = requests.post(query_endpoint, headers=headers, json=query_payload)

if response.status_code in [200, 202]:
    # Extracting and printing query results
    query_results = response.json()
    print("Query executed successfully. Results:")
    print(query_results)
else:
    print("Failed to execute query. Check your query and datasource ID.")
    print(response)
    print(response.json())

# Note: Adjust the query payload as per your specific requirements.
```
